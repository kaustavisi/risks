```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Fetching package health
An organization https://synk.io provides python package health score by doing comprehensive analysis of the packages from multiple aspects. Currently, I'm mining data from their webpage. This is purely for research purpose & for any serious business use, one should get API access from the organization. This is not an ethical practice to mine data like this. 

```python
def get_package_health(pack):
    url = "https://snyk.io/advisor/python/" + pack
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    ## Get Security, Popularity, Maintenance & Community stats 
    d = soup.findAll('div', class_='package-extra')[0].\
        findAll('div', class_='health')[0].\
        findAll('div', class_='vue--pill--uppercase')
    pd_df = pd.DataFrame({'Package': [pack], 
                         'Score': [soup.findAll('title')[2].text.split(":")[1]],
                         'Security': [d[0].text],
                         'Popularity': [d[1].text],
                         'Maintenance': [d[2].text],
                         'Community': [d[3].text]
                        })
    ## Get other additional statistics of the package
    dd = [x.text.split('\n')[:2] for x in soup.findAll('div', class_='stats-item')]
    df = {}
    for x in dd:
        df[x[0].strip()] = x[1].strip()
    pd_df2 = pd.DataFrame(df, index=['i',])
    
    ## Merge both stat & return
    return(pd_df.merge(pd_df2, how = 'cross'))
```

```python
packs = ['numpy', 'pandas', 'qiling', 'wfdb', 'netflix']
d = [get_package_health(pack) for pack in packs]
pd.concat(d)
```
